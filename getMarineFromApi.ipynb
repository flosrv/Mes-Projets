```python
from siphon.simplewebservice.ndbc import NDBC
import pandas as pd
```

Get Marine Data from API

```python
df_marine = NDBC.realtime_observations('51101')
```

Check first rows

```python
df_marine.head()
```

Rename Columns + Delete Useless Ones

```python
df_marine.rename(columns={'wind_direction': 'Wind Dir (°)', 'wind_speed':'Wind Speed (ms)',
                          'wind_gust':'Wind Gusts 1h (ms)','wave_height':'Wave Height (m)','dominant_wave_period':'Dominant Wave Period (s)',
                          'average_wave_period':'Avg Wave Period (s)', 'dominant_wave_direction': 'Dominant Wave Direction',
                          'air_temperature':'Air Temperature','water_temperature':'Water Temperature','time': 'Datetime','pressure':'Pressure'}, inplace=True)


df_marine = df_marine.drop(columns={'dewpoint', 'visibility','3hr_pressure_tendency','water_level_above_mean'})
```

Check Head again

```python
df_marine.head(5)
```

Check Nombres valeurs vides / Nombre total de lignes

```python
print(df_marine.isnull().sum(),df_marine.shape)
```

Check Data Types

```python
df_marine.dtypes
```

Delete Rows With Empty Values in Data Columns

```python
# Spécifier les colonnes à considérer pour la suppression des lignes
columns_to_check = ['Wind Dir (°)', 'Wind Speed (ms)', 'Wind Gusts 1h (ms)', 'Wave Height (m)', 'Dominant Wave Period (s)', 
                    'Avg Wave Period (s)', 'Dominant Wave Direction', 'Pressure', 'Air Temperature', 'Water Temperature']

# Supprimer les lignes avec des valeurs nulles dans les colonnes spécifiées
df_marine.dropna(subset=columns_to_check, inplace=True)

# Afficher le DataFrame après la suppression des lignes
df_marine.isnull().sum()
```
